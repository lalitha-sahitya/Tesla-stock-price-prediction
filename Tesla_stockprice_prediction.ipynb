```python
import opendatasets as od
import pandas as pd
from prophet import Prophet
import plotly.express as px
```

```python
import plotly.io as pio
pio.renderers.default='colab'
```

```python
od.download('https://www.kaggle.com/datasets/harshsingh2209/tesla-stock-pricing-20172022')
```

```python
df=pd.read_csv('/content/tesla-stock-pricing-20172022/TSLA.csv')
```

```python
px.line(df,x='Date',y='Close')
```

```python
px.bar(df,x='Date',y='Volume')
```

```python
px.box(df,y='Close')
```

```python
cols=['Date','Close']
```

```python
new_df=pd.DataFrame(df, columns=cols)
```

```python
new_df=new_df.rename(columns={'Date':'ds','Close':'y'})
```

```python
new_df
```

```python
m=Prophet()
m.fit(new_df)
```

```python
future=m.make_future_dataframe(periods=365)
forecast=m.predict(future)
```

```python
forecast
```

```python
px.line(forecast, x='ds', y='yhat')
```

```python
forecast_new=forecast.tail(365)
```

```python
forecast_new
```

```python
forecast_new_df=pd.DataFrame(forecast_new, columns=['ds','yhat'])
```

```python
forecast_new_df
```

```python
forecast_new_df.to_csv('forecast.csv')
```
